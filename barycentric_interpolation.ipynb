```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BarycentricInterpolator
```

```python
# Given data points
x_points = np.array([1, 3, 4])
y_points = np.array([2, 8, 3])

# Create the Barycentric Interpolator
interpolator = BarycentricInterpolator(x_points, y_points)

# Evaluate at x = 2
x_eval = 2
y_eval = interpolator(x_eval)

# Generate smooth curve for visualization
x_smooth = np.linspace(0, 5, 100)
y_smooth = interpolator(x_smooth)

# Plot the interpolation
plt.figure(figsize=(8, 5))
plt.plot(x_smooth, y_smooth, label="Barycentric Interpolation", color='blue')
plt.scatter(x_points, y_points, color='red', label="Given Points", zorder=3)
plt.scatter(x_eval, y_eval, color='green', label=f"Interpolated y at x={x_eval}: {y_eval:.2f}", zorder=3)
plt.axvline(x_eval, color='gray', linestyle='dashed', alpha=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Barycentric Interpolation with Three Points")
plt.legend()
plt.grid(True)
plt.show()

# Return the interpolated value
y_eval
```

```python
np.random.seed(42)  # Fixing the random seed for reproducibility

# Parameters

N = 512  # Total number of points
num_samples = 50  # Number of sample points for interpolation
x = np.linspace(0, 1, N)  # Full range
sample_indices = np.sort(np.random.choice(N, num_samples, replace=False))
sample_x = x[sample_indices]
sample_y = np.sin(2 * np.pi * sample_x)  # Sampled function values

# Barycentric interpolation

interpolator = BarycentricInterpolator(sample_x, sample_y)
interpolated_y = interpolator(x)

# Plotting

plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(2 * np.pi * x), label='Original Function', linestyle='dashed')
plt.plot(sample_x, sample_y, 'ro', label='Sampled Points')
plt.plot(x, interpolated_y, label='Barycentric Interpolation')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Function Reconstruction using Barycentric Interpolation')
plt.legend()
plt.show()
```
